## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Create an overview table that summarizes key metrics about each school 

school_data_complete['Pass Math'] = school_data_complete["math_score"] >70
school_data_complete['Pass Reading'] = school_data_complete["reading_score"] >70

#group by school name and type
grouped = school_data_complete.groupby(['school_name','type'])


#count the number of students and sum the budget
grouped = grouped.agg({'Student ID':'count', 'budget': 'median','math_score': 'mean', 'reading_score':'mean', 
                       'Pass Math':'sum', 'Pass Reading':'sum'})
grouped.head(5)
```

```python
# Percentage Student/Budget
grouped["Per Student Budget"] = grouped["budget"] / grouped["Student ID"]

#Percentage of Students who passed math
grouped["% Passed Math"] = grouped["Pass Math"] / grouped["Student ID"]

#Percentage of Students who passed reading
grouped["% Passed Reading"] = grouped["Pass Reading"] / grouped["Student ID"]

#Overall passing rate
grouped["Overall Passing Rate"] = (grouped["% Passed Math"] + grouped["% Passed Reading"])/2



#rename columns
grouped = grouped.rename(columns={'school_name':'School Name', "type":"School Type", "Student ID": "Total Students", 
                                  "budget": "Total School Budget", "math_score": "Average Math Score", 
                                  "reading_score":"Average Reading Score"})
grouped.head(10)
```

```python
type(grouped)
grouped.index
GroupedReset = grouped.reset_index()

GroupedReset = GroupedReset.rename(columns={'school_name':'School Name', "type":"School Type"})

GroupedReset.head(10)
```

```python
#Create a dataframe to hold the above results

SchoolSummDF = GroupedReset[["School Name", "School Type", "Total Students", "Total School Budget",
                             "Per Student Budget", "Average Math Score", "Average Reading Score", 
                             "% Passed Math", "% Passed Reading", "Overall Passing Rate"]]
SchoolSummDF.head(10)

#SchoolSummDF = grouped.loc['school_name', 'type', 'Total Students', 'Total School Budget','Per Student Budget', 'Average Math Score', 
                       #'Average Reading Score', '% Passed Math', '% Passed Reading', 'Overall Passing Rate']
#SchoolSummDF.head(10)
```
